```python
def Distribution_check(dataframes, boxplot_percentile, groups, colname):
    
    """
    Creates histograms and boxplots from the given dataframe/s

    - dataframes: list of dataframes which distribution you want to check
    - randoms: list of dataframes which mean you want to considered as random sample
    - boxplot_percentile: list. Interval € [0,100] in which you want to limit the boxplots (in case of exponential)
    - groups: list. Names of the dataframes you are analycing
    - colname: list. Columns wich the scores are located. Dataframes equivalent scores must share conames
    """

    # Figure definition
    fig_size = plt.rcParams["figure.figsize"] = [24,18]
    n_score = len(colname) #nº score
    n_dataframes = len(dataframes)

    # Recolector/plotting loop
    for n_s in range(n_score): #loop through scores  
        scores = []
        perc_scores = []
        
        ## score recolector, depende del tipo de datos
        for n_df in range(n_dataframes): #loop through dataframes
            # from list of predictions of a similar nº of random proteins
            if isinstance(dataframes[n_df], list):
                df = dataframes[n_df][random.randrange(0,len(dataframes[n_df]))] # se selecciona uno aleatoriamente

            else:
                df = dataframes[n_df]
                
            score = df[colname[n_s]].dropna().values
            ### general/hist
            positive = False if min(score)<0 else True  #pythonic if/else 
            scores = np.log(abs(score)+1)
            ### boxplot
            raw_s = df[colname[n_s]].dropna().values
            perc = np.percentile(raw_s, boxplot_percentile)
            perc_scores.append(raw_s[(raw_s>perc[0])*(raw_s<perc[1])])

            
            ## Plot
            ### hist
            plt.subplot(n_score, 2 , n_s*2+1)
            if positive == False:
                scores = [ -x for x in scores ] #change score sign
                #plt.hist(scores, label=groups, bins=50, density=True)
                sns.kdeplot(scores, label=groups[n_df], shade= False, alpha = 1) # shade por si lo quieres relleno
                plt.xticks(plt.xticks()[0],-np.around(np.exp(-plt.xticks()[0])-1,2))
                


            else: 
                #plt.hist(scores, label=groups, bins=50, density=True)
                sns.kdeplot(scores, label=groups[n_df], shade= False, alpha = 1) # shade por si lo quieres relleno
                plt.xticks(plt.xticks()[0],np.around(np.exp(plt.xticks()[0])-1,2))
        plt.xlim((perc))
        plt.legend('')
        #plt.legend(fontsize=12, loc = "best")
        plt.xlabel(colname[n_s])
        plt.ylabel("density")
        
            
        ###hist mkup & random    
        
        ### boxplot
        plt.subplot(n_score, 2 , n_s*2+2)
        sns.violinplot(data = perc_scores)
        plt.title("Score Distribution", fontweight="bold")

        plt.xlabel("Group", fontweight="semibold")
        plt.ylabel(colname[n_s], fontweight="semibold")
        plt.xticks(plt.xticks()[0],groups)
        plt.plot()
```

```python
from collections import Counter
def Compute_subsets(a, b):
    '''
    Given two set or Counter objects, computes the sizes of (a & ~b, b & ~a, a & b).
    Returns the result as a tuple.
    (only in a, only in b, common)
    '''
    if not (type(a) == type(b)):
        raise ValueError("Both arguments must be of the same type")
    set_size = len if type(a) != Counter else lambda x: sum(x.values())   # We cannot use len to compute the cardinality of a Counter
    return (set_size(a - b), set_size(b - a), set_size(a & b))

def Unique_miRs_subset (og_datasets, positive, threshold_value, threshold_fields):
    """
    IMPORTANTE MANTENER EL ORDEN DE og_datasets
    """
    # input asociation
    general,trxG, PcG = og_datasets

    # threshold aplication
    if positive == True:
        general = general.sort_values(by = threshold_fields, ascending = False).iloc[:int(general.shape[0]*threshold_value)]
        trxG = trxG[trxG[threshold_fields[0]] > min(general[threshold_fields[0]])]
        PcG = PcG[PcG[threshold_fields[0]] > min(general[threshold_fields[0]])]
        
    elif positive == False:
        general = general.sort_values(by = threshold_fields, ascending = True).iloc[:int(general.shape[0]*threshold_value)]
        trxG = trxG[trxG[threshold_fields[0]] < max(general[threshold_fields[0]])]
        PcG = PcG[PcG[threshold_fields[0]] < max(general[threshold_fields[0]])]
    
    # (only in a, only in b, common) calculation
    ## PcG and trxG
    subsets = Compute_subsets(set(trxG["miRNA family"].drop_duplicates()),
                              set(PcG["miRNA family"].drop_duplicates()))
    
    ## Random (specific for PcG and trxG)
    r_PcG_subsets = np.zeros((10,3))
    r_trxG_subsets = np.zeros((10,3))
    
    for n in range(10): 
        #j# Se mantiene siempre el orden de los subsets
        r_trxG_subsets[n,:] = Compute_subsets(set(trxG["miRNA family"].drop_duplicates()),
                                             set(general.sample(n=PcG.shape[0])["miRNA family"].drop_duplicates()))
        
        r_PcG_subsets[n,:] = Compute_subsets(set(general.sample(n=trxG.shape[0])["miRNA family"].drop_duplicates()),
                                            set(PcG["miRNA family"].drop_duplicates()))
    
    return(subsets, r_trxG_subsets, r_PcG_subsets)
```

```python
# loop ploteador quitador
def Dif_plotter (miRs,scores, step):
    fig_size = plt.rcParams["figure.figsize"] = [12,12]
    n_pred = miRs.shape[0]
    
    while n_pred >0:
        ## update df taking last row
        miRs=miRs[:n_pred]
        scores=scores[:n_pred]

        ## Check prints
        #print(miRs.shape)        # checks df new shape
        #print(miRs.values[-1])   # check df last row

        ## overlap PcG w trxG
        #j# (only_a, only_b, both)
        subsets = Compute_subsets(set(miRs["pcg"].drop_duplicates()), 
                                  set(miRs["trxg"].drop_duplicates()))
        r_PcG_subsets = np.zeros((10,3))
        r_trxG_subsets = np.zeros((10,3))
        for n in range(10):

            r_PcG_subsets[n,:] = Compute_subsets(set(miRs["pcg"].drop_duplicates()), 
                                                 set(miRs["Random_%i"%(n)].drop_duplicates()))

            r_trxG_subsets[n,:] = Compute_subsets(set(miRs["trxg"].drop_duplicates()), 
                                                 set(miRs["Random_%i"%(n)].drop_duplicates()))

        ## plots
        ### normal
        plt.subplot(211)
        plt.plot(n_pred,(subsets[2]/(subsets[0]+subsets[2])), c='C3',marker ='x',ms=5) #normal PcG
        plt.plot(n_pred,(subsets[2]/(subsets[1]+subsets[2])), c='C2',marker='x',ms=5) #normal trxG

        ### random
        plt.errorbar(n_pred, np.mean(r_PcG_subsets[:,2]/(r_PcG_subsets[:,0]+r_PcG_subsets[:,2])),
                     np.std(r_PcG_subsets[:,2]/(r_PcG_subsets[:,0]+r_PcG_subsets[:,2])), 
                     linestyle='None', marker='.', ms=7, color = "C3", alpha = 0.5)

        plt.errorbar(n_pred, np.mean(r_trxG_subsets[:,2]/(r_trxG_subsets[:,0]+r_trxG_subsets[:,2])),
                     np.std(r_trxG_subsets[:,2]/(r_trxG_subsets[:,0]+r_trxG_subsets[:,2])), 
                     linestyle='None', marker='.',ms=7, color = "C2", alpha = 0.5)

        ## plots scores
        plt.subplot(212)
        plt.plot(n_pred, scores["pcg"].values[-1], c='C3',marker='x',ms=5)
        plt.plot(n_pred, scores["trxg"].values[-1], c='C2',marker='x',ms=5)
        plt.errorbar(n_pred, np.mean(scores.loc[:,scores.columns.str.match('Random_')].dropna(axis = 1, how = "any").values[-1]),
                     np.std(scores.loc[:,scores.columns.str.match('Random_')].values[-1]), linestyle='None', marker='*',ms=7, color = "C1", alpha = 0.5)

        n_pred-=step
    # makeup
    ## miRNAs
    plt.subplot(211)
    ### legend
    PcG_leg = mlines.Line2D([], [], color='C3', marker='x',lw = 0,  markersize=5, label='PcG % overlap with trxG')
    trxG_leg = mlines.Line2D([], [], color='C2', marker='x',lw = 0, markersize=5, label='trxG % overlap with PcG')
    random_PcG_leg = mlines.Line2D([], [], color='C3', marker='.',lw = 0, markersize=7, label='PcG % overlap with Random')
    random_trxG_leg = mlines.Line2D([], [], color='C2', marker='.',lw = 0, markersize=7, label='trxG % overlap with Random')

    plt.legend(handles = [PcG_leg,trxG_leg, random_PcG_leg, random_trxG_leg], fontsize=12, loc = 'lower right')
    ### other
    plt.title("miRNA Overlap among groups", fontweight="bold")
    plt.xlabel("nº of predictions considered", fontweight="semibold")
    plt.ylabel("% of shared miRNAs", fontweight="semibold")

    ## score
    plt.subplot(212)
    ### legend
    PcG_leg = mlines.Line2D([], [], color='C3', marker='x',lw = 0 ,ms=5, label='PcG')
    trxG_leg = mlines.Line2D([], [], color='C2', marker='x',lw = 0 ,ms=5, label='trxG')
    random_leg = mlines.Line2D([], [], color='C1', marker='*',lw = 0,ms=7 , label='Random')
    plt.legend(handles = [PcG_leg, trxG_leg, random_leg], fontsize=12, loc = 'lower right')
    plt.tight_layout()
    ### other
    plt.title("Minimum Score", fontweight="bold")
    plt.xlabel("Order", fontweight="semibold")
    plt.ylabel("Score", fontweight="semibold")


    
```

```python
from collections import Counter
def Compute_subsets(a, b):
    '''
    Given two set or Counter objects, computes the sizes of (a & ~b, b & ~a, a & b).
    Returns the result as a tuple.
    (only in a, only in b, common)
    '''
    if not (type(a) == type(b)):
        raise ValueError("Both arguments must be of the same type")
    set_size = len if type(a) != Counter else lambda x: sum(x.values())   # We cannot use len to compute the cardinality of a Counter
    return (set_size(a - b), set_size(b - a), set_size(a & b))

def Processing (og_datasets, positive, threshold_value, threshold_fields):
    """
    IMPORTANTE MANTENER EL ORDEN DE og_datasets
    """
    # input asociation
    general,trxG, PcG = og_datasets

    # threshold aplication
    if positive == True:
        general = general.sort_values(by = threshold_fields, ascending = False).iloc[:int(general.shape[0]*threshold_value)]
        trxG = trxG[trxG[threshold_fields[0]] > min(general[threshold_fields[0]])]
        PcG = PcG[PcG[threshold_fields[0]] > min(general[threshold_fields[0]])]
        
    elif positive == False:
        general = general.sort_values(by = threshold_fields, ascending = True).iloc[:int(general.shape[0]*threshold_value)]
        trxG = trxG[trxG[threshold_fields[0]] < max(general[threshold_fields[0]])]
        PcG = PcG[PcG[threshold_fields[0]] < max(general[threshold_fields[0]])]
    
    # (only in a, only in b, common) calculation
    ## PcG and trxG
    subsets = Compute_subsets(set(trxG["miRNA family"].drop_duplicates()),
                              set(PcG["miRNA family"].drop_duplicates()))
    
    ## Random (specific for PcG and trxG)
    r_PcG_subsets = np.zeros((10,3))
    r_trxG_subsets = np.zeros((10,3))
    
    for n in range(10): 
        #j# Se mantiene siempre el orden de los subsets
        r_trxG_subsets[n,:] = Compute_subsets(set(trxG["miRNA family"].drop_duplicates()),
                                             set(general.sample(n=PcG.shape[0])["miRNA family"].drop_duplicates()))
        
        r_PcG_subsets[n,:] = Compute_subsets(set(general.sample(n=trxG.shape[0])["miRNA family"].drop_duplicates()),
                                            set(PcG["miRNA family"].drop_duplicates()))
    
    return(subsets, r_trxG_subsets, r_PcG_subsets)
```
