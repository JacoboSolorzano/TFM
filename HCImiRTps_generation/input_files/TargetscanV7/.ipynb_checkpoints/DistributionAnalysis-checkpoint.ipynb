```python
import csv
from difflib import SequenceMatcher
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 1000)
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
import matplotlib.lines as mlines
```

# Analisis de la distribiución
----------
## Procesamiento y funciones

```python
# Importamos las tablas de genes y predicción
UTR3_info = pd.read_csv(filepath_or_buffer='PcGtrxG_UTRinfo_noncolapsed.csv',index_col= 0, sep = ",")
miR_targ = pd.read_csv(filepath_or_buffer='Summary_Counts.all_predictions.txt', sep = "\t")

# Generamos tabla de genes reducida (quitamos columnas)
UTR3_info_reduced = UTR3_info[["Group", "external_gene_name", "gene_id","ensembl_gene_id", "ensembl_transcript_id"]].drop_duplicates()

# Juntamos la tabla de genes reducida con la predicción asi filtrandola
miRtarg_info = UTR3_info_reduced.join(miR_targ.set_index("Transcript ID"), on = "ensembl_transcript_id")

# Eliminamos NAs (al fusionar tablas genes sin targets se incorporan rellenando con NAs)
miRtarg_info =  miRtarg_info.dropna(axis = 0, subset= ["Species ID"]) #eliminamos aquellas filas que no tienen target

# Generamos datasets dividiendo por el grupo
miRtarg_trxG = miRtarg_info[miRtarg_info['Group'] == "trxG"]
miRtarg_PcG = miRtarg_info[miRtarg_info['Group'] == "PcG"]
```

```python
from collections import Counter
def Compute_subsets(a, b):
    '''
    Given two set or Counter objects, computes the sizes of (a & ~b, b & ~a, a & b).
    Returns the result as a tuple.
    (only in a, only in b, common)
    '''
    if not (type(a) == type(b)):
        raise ValueError("Both arguments must be of the same type")
    set_size = len if type(a) != Counter else lambda x: sum(x.values())   # We cannot use len to compute the cardinality of a Counter
    return (set_size(a - b), set_size(b - a), set_size(a & b))

def Processing (og_datasets, positive, threshold_value, threshold_fields):
    """
    IMPORTANTE MANTENER EL ORDEN DE og_datasets
    """
    # input asociation
    general,trxG, PcG = og_datasets

    # threshold aplication
    if positive == True:
        general = general.sort_values(by = threshold_fields, ascending = False).iloc[:int(general.shape[0]*threshold_value)]
        trxG = trxG[trxG[threshold_fields[0]] > min(general[threshold_fields[0]])]
        PcG = PcG[PcG[threshold_fields[0]] > min(general[threshold_fields[0]])]
        
    elif positive == False:
        general = general.sort_values(by = threshold_fields, ascending = True).iloc[:int(general.shape[0]*threshold_value)]
        trxG = trxG[trxG[threshold_fields[0]] < max(general[threshold_fields[0]])]
        PcG = PcG[PcG[threshold_fields[0]] < max(general[threshold_fields[0]])]
    
    # (only in a, only in b, common) calculation
    ## PcG and trxG
    subsets = Compute_subsets(set(trxG["miRNA family"].drop_duplicates()),
                              set(PcG["miRNA family"].drop_duplicates()))
    
    ## Random (specific for PcG and trxG)
    r_PcG_subsets = np.zeros((100,3))
    r_trxG_subsets = np.zeros((100,3))
    
    for n in range(100): 
        #j# Se mantiene siempre el orden de los subsets
        r_trxG_subsets[n,:] = Compute_subsets(set(trxG["Representative miRNA"].drop_duplicates()),
                                             set(general.sample(n=PcG.shape[0])["Representative miRNA"].drop_duplicates()))
        
        r_PcG_subsets[n,:] = Compute_subsets(set(general.sample(n=trxG.shape[0])["Representative miRNA"].drop_duplicates()),
                                            set(PcG["Representative miRNA"].drop_duplicates()))
    
    return(subsets, r_trxG_subsets, r_PcG_subsets)
        
```

## Basado en CWCS

```python
# function calling and plotting loop
for n in np.linspace(0,1,101)[1:]:
    ## function calling (only in trx, only in Pc, common)
    subsets,r_trxG_subsets,r_PcG_subsets = Processing([miR_targ, miRtarg_trxG, miRtarg_PcG], False, n,
                                                      ["Cumulative weighted context++ score", "Total context++ score"])

    ## plotting
    ### trxG
    plt.subplot(121)
    plt.plot(n,(subsets[0]/(subsets[0]+subsets[2])), 'rx') #normal
    plt.errorbar(n, np.mean(r_trxG_subsets[0]/(r_trxG_subsets[0]+r_trxG_subsets[2])),
                 np.std(r_trxG_subsets[0]/(r_trxG_subsets[0]+r_trxG_subsets[2])), linestyle='None', marker='.', color = "b")
    
    ### Pcg
    plt.subplot(122)
    plt.plot(n,(subsets[1]/(subsets[1]+subsets[2])), 'rx')
    plt.errorbar(n, np.mean(r_PcG_subsets[1]/(r_PcG_subsets[1]+r_PcG_subsets[2])),
                 np.std(r_PcG_subsets[1]/(r_PcG_subsets[1]+r_PcG_subsets[2])), linestyle='None', marker='.', color = "b")

    
# makeup
## PcG
plt.subplot(121)
### legend
trx_leg = mlines.Line2D([], [], color='red', marker='x',lw = 0,  markersize=10, label='vs. PcG')
random_leg = mlines.Line2D([], [], color='blue', marker='.',lw = 0, markersize=15, label='Random')
plt.legend(handles = [trx_leg, random_leg])
### other
plt.title("trxG % of unique miRNAs")
plt.xlabel("% of top predictions")
plt.ylabel("% of unique miRNAs")

## trxG
plt.subplot(122)
### legend
PcG_leg = mlines.Line2D([], [], color='red', marker='x',lw = 0, markersize=10, label='vs. trxG')
random_leg = random_leg
plt.legend(handles = [PcG_leg, random_leg])
### other
plt.title("PcG % of unique miRNAs")
plt.xlabel("% of top predictions")
plt.ylabel("% of unique miRNAs")

plt.plot()
```
