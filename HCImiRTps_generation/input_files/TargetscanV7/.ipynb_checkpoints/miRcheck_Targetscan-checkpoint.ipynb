### GO de los miR no comunes (PcG/trxG) via GOATools
codigo adaptado de https://github.com/tanghaibao/goatools/blob/master/notebooks/goea_nbt3102.ipynb

En este haremos GO de los genes que estan siendo regulados por los miR que regulan unicamente a uno u a otro grupo, tomando como base todos lo genes qure aparecen en el general

GOATOOLS solo trabaja con ensmblIDs, dato que no provee la prediccion. 
    1º Convertimos basándonos en transcript id unicos
    2º eliminamos entrezids duplicados
    3º eliminamos NAs
    
Por el camino perdemos targets, pero perdemos los mismos en control que en TrxG/PcG asique no pasa nada

**Funciones Necesarias**


```python
import mygene
def Get_ids(pd_series):
    """
    Desde una Serie de Pandas de Ensmbl Transcript IDs devuelve la lista de Entrez IDs 
    Necesaria para realizar el GOATools
    """
    mg = mygene.MyGeneInfo()
    mygene_result = mg.querymany(pd_series, scopes="ensembl.transcript", species='fruitfly')
   
    ids = np.zeros((len(mygene_result)))
    symbols = pd.Series(data = np.zeros((len(mygene_result))))
    print(mygene_result[0])


    for i in range(len(mygene_result)):
        try:
            ids[i] = mygene_result[i]['entrezgene']

        except:
            ids[i] = None

        try:
            symbols[i] = mygene_result[i]['symbol']

        except:
            symbols[i]= None



    return ids, symbols
```

**Codigo necesario para realizar GO (seleccionando taxid etc)**

```python
from goatools.base import download_go_basic_obo , download_ncbi_associations

obo_fname = download_go_basic_obo()
fin_gene2go = download_ncbi_associations()

from goatools.obo_parser import GODag
# Carga ontologia

obodag = GODag("go-basic.obo")


from __future__ import print_function
from goatools.anno.genetogo_reader import Gene2GoReader
               
# Read NCBI's gene2go. Store annotations in a list of namedtuples
objanno = Gene2GoReader(fin_gene2go, taxids=[7227]) #7227 es Dmel

# Get namespace2association where:
#    namespace is:
#        BP: biological_process               
#        MF: molecular_function
#        CC: cellular_component
#    assocation is a dict:
#        key: NCBI GeneID
#        value: A set of GO IDs associated with that gene
ns2assoc = objanno.get_ns2assc()

for nspc, id2gos in ns2assoc.items():
    print("{NS} {N:,} annotated Dmel genes".format(NS=nspc, N=len(id2gos)))
```

**Creacion lista de genes de base/importacion**

Como base todos los genes que figuran en documento de predicción original (miR_Targ)

```python
# Opcion con todos los protcod genes
from goatools.test_data.genes_NCBI_7227_ProteinCoding import GENEID2NT as GeneID2nt_Dmel
# Mejor usar como base todos los genes de la prediccion global
GeneIDglobal_Dmel, symbolcheck = Get_ids(miR_targ["Transcript ID"].drop_duplicates())# eliminamos transcritos duplicados
GeneIDglobal_Dmel = np.unique(GeneIDglobal_Dmel) # eliminamos ids duplicados
GeneIDglobal_Dmel = GeneIDglobal_Dmel[~np.isnan(GeneIDglobal_Dmel)] #eliminamos NAs
```

**GOEA en sí**
- PcG

Los miR que afectan a PcG y no a trxG

```python
from goatools.goea.go_enrichment_ns import GOEnrichmentStudyNS

# Creacion objeto GO
goeaobj = GOEnrichmentStudyNS(
        list(GeneIDglobal_Dmel), # List of Dmel protein-coding genes seria asi -> GeneID2nt_Dmel.keys()
        ns2assoc, # geneid/GO associations
        obodag, # Ontologies
        propagate_counts = False,
        alpha = 0.05, # default significance cut-off
        methods = ['fdr_bh']) # defult multipletest correction method

# Generacion lista de geneIDS regulados pòr los miR que regulan solo a PcG

GOmiR_PcG = []
Pc_miRlist = exclusive_miRs["only_PcG"].dropna().values

for index in range(Pc_miRlist.shape[0]):
    GOmiR_PcG = GOmiR_PcG + list(miR_targ[miR_targ["Representative miRNA"] == Pc_miRlist[index]]["Transcript ID"].values)
    
GOmiR_PcG = list(dict.fromkeys(GOmiR_PcG)) #para eliminar transcritos duplicados
GOmiR_PcG, symbolcheck = Get_ids(GOmiR_PcG)
GOmiR_PcG = np.unique(GOmiR_PcG) # para eliminar geneids duplicados
GOmiR_PcG = GOmiR_PcG[~np.isnan(GOmiR_PcG)] # para eliminar Nas

# Realizacion del GO en si
goea_results_all = goeaobj.run_study(list(GOmiR_PcG))
goea_results_sig = [r for r in goea_results_all if r.p_fdr_bh < 0.05]

# Recoleccion resultados
goeaobj.wr_xlsx("GOmiR_PcG.xlsx", goea_results_sig)
goeaobj.wr_txt("GOmiR_PcG.txt", goea_results_sig)

# Obtencion de imagenes
from goatools.godag_plot import plot_gos, plot_results, plot_goid2goobj

plot_results("GOmiR_PcG_{NS}.png", goea_results_sig)
```

-  trxG

```python
from goatools.goea.go_enrichment_ns import GOEnrichmentStudyNS

# Creacion objeto GO
goeaobj = GOEnrichmentStudyNS(
        list(GeneIDglobal_Dmel), # List of Dmel protein-coding genes seria asi -> GeneID2nt_Dmel.keys()
        ns2assoc, # geneid/GO associations
        obodag, # Ontologies
        propagate_counts = False,
        alpha = 0.05, # default significance cut-off
        methods = ['fdr_bh']) # defult multipletest correction method

# Generacion lista de geneIDS regulados pòr los miR que regulan solo a trxG

GOmiR_trxG = []
trx_miRlist = exclusive_miRs["only_trxG"].dropna().values

for index in range(trx_miRlist.shape[0]):
    GOmiR_trxG = GOmiR_trxG + list(miR_targ[miR_targ["Representative miRNA"] == trx_miRlist[index]]["Transcript ID"].values)
    
GOmiR_trxG = list(dict.fromkeys(GOmiR_trxG)) #para eliminar transcritos duplicados
GOmiR_trxG, symbolcheck = Get_ids(GOmiR_trxG)
GOmiR_trxG = np.unique(GOmiR_trxG) # para eliminar geneids duplicados
GOmiR_trxG = GOmiR_trxG[~np.isnan(GOmiR_trxG)] # para eliminar Nas

# Realizacion del GO en si
goea_results_all = goeaobj.run_study(list(GOmiR_trxG))
goea_results_sig = [r for r in goea_results_all if r.p_fdr_bh < 0.05]

# Recoleccion resultados
goeaobj.wr_xlsx("GOmiR_trxG.xlsx", goea_results_sig)
goeaobj.wr_txt("GOmiR_trxG.txt", goea_results_sig)

# Obtencion de imagenes
from goatools.godag_plot import plot_gos, plot_results, plot_goid2goobj

plot_results("GOmiR_trxG_{NS}.png", goea_results_sig)
```

### Procesamiento
Generamos una tabla con filas como diferentes miRs y columnas como conteo de genes a los que afecta en cada uno de los grupos, que luego normalizamos dividiendo entre el total de genes del grupo

```python
# Creamos una tabla que contiene todos los miRNAs con su frecuencia total y para cada grupo
## Obtencion miR individuales
miR_freq = miRtarg_info["Representative miRNA"].drop_duplicates()

## Creacion tabla vacia (para ahorrar espacio meto columna que luego modifico)
miR_freq = pd.concat([miR_freq, miR_freq,miR_freq, miR_freq, miR_freq, miR_freq, miR_freq], axis = 1)
miR_freq.columns = ["Representative miRNA","genome_freq", "total_freq", "PcG_freq", "trxG_freq", "Hox_freq", "recruiter_freq" ]
miR_freq.reset_index(drop= True, inplace= True) # permite mejor acceso para rellenar

## rellenado de la tabla
for index in range(len(miR_freq["Representative miRNA"])):
    # suma de todas las entradas de tabla X que en el campo de miR_accession tengan el correspondiente al index en miR_freq
    miR_freq["total_freq"][index] = sum(miRtarg_info["Representative miRNA"] == miR_freq["Representative miRNA"][index])
    miR_freq["genome_freq"][index] = miR_targ[miR_targ["Representative miRNA"]== miR_freq["Representative miRNA"][index]].shape[0] # cuantas dianas se han detectado en total para ese mIR
    miR_freq["PcG_freq"][index] = sum(miRtarg_PcG["Representative miRNA"] == miR_freq["Representative miRNA"][index])
    miR_freq["trxG_freq"][index] = sum(miRtarg_trxG["Representative miRNA"] == miR_freq["Representative miRNA"][index])
    miR_freq["Hox_freq"][index] = sum(miRtarg_Hox["Representative miRNA"] == miR_freq["Representative miRNA"][index])
    miR_freq["recruiter_freq"][index] = sum(miRtarg_recruiter["Representative miRNA"] == miR_freq["Representative miRNA"][index])

## sanity check
### Los miR de PcG trxG recruiters y Hox = total
print(all(miR_freq["total_freq"].values == miR_freq["PcG_freq"].values + miR_freq["trxG_freq"].values + miR_freq["Hox_freq"].values + miR_freq["recruiter_freq"].values))

### La suma denmiR de PcG corresponde a la longitud de las listas de los miR que le afectan
print(sum(miR_freq["trxG_freq"].values) == len(miRtarg_trxG.values) and 
      sum(miR_freq["PcG_freq"].values) == len(miRtarg_PcG.values) and
      sum(miR_freq["Hox_freq"].values) == len(miRtarg_Hox.values) and
      sum(miR_freq["recruiter_freq"].values) == len(miRtarg_recruiter.values))

# Normalización
miR_freq["PcG_freq"] = miR_freq["PcG_freq"] / UTR3_info_reduced[UTR3_info_reduced['Group'] == "PcG"]["external_gene_name"].drop_duplicates().shape[0]
miR_freq["trxG_freq"] = miR_freq["trxG_freq"] / UTR3_info_reduced[UTR3_info_reduced['Group'] == "trxG"]["external_gene_name"].drop_duplicates().shape[0]
miR_freq["Hox_freq"] = miR_freq["Hox_freq"] / UTR3_info_reduced[UTR3_info_reduced['Group'] == "Hox"]["external_gene_name"].drop_duplicates().shape[0]
miR_freq["recruiter_freq"] = miR_freq["recruiter_freq"] / UTR3_info_reduced[UTR3_info_reduced['Group'] == "recruiter"]["external_gene_name"].drop_duplicates().shape[0]
miR_freq["genome_freq"] = miR_freq["genome_freq"] / miR_targ["Gene Symbol"].drop_duplicates().shape[0]
```

```python
# colocacion de miRBase_acc como index y formateo de los datos para heatmap
miR_freq = miR_freq.set_index(keys=["Representative miRNA"]).astype('float64')
```

```python
def Old_Distribution_check(dataframes,groups, colname):
    """
    """
    fig_size = plt.rcParams["figure.figsize"] = [20,10]
    n_scores = len(colname) # nº scores
    n_dataframes = len(dataframes)
    stats = np.zeros((n_dataframes*n_scores, 5))
    stats_index = []
    
    for n_df in range(n_dataframes):
        df = dataframes[n_df]
        
        for n_s in range(n_scores):
            stats_index.append(groups[n_df] + "_" + colname[n_s])
            
            scores = df[colname[n_s]].dropna().values
            stats[n_df*3 + n_s, :] = np.array([np.mean(scores),
                                               np.median(scores),
                                               np.max(scores),
                                               np.min(scores),
                                               scores.shape[0]])
            
            
            plt.subplot(n_dataframes ,n_scores , n_df*3 + n_s +1)
            plt.hist(scores, bins=1000)
            
            if n_df*3 + n_s +1 <=3:    #titulo col
                plt.title(colname[n_s])
            if (n_df*3 + n_s)%3 == 0:  #titulo row (cambiar cuando averigue como va)
                plt.ylabel(groups[n_df])
                
    plt.tight_layout()
    plt.plot()
        
    
                
    return pd.DataFrame(data = stats, 
                 columns= ["Mean", "Median", "max", "min", "nº of instances"],
                index = stats_index)

    
        
Old_Distribution_check([miR_targ, miRtarg_PcG, miRtarg_trxG],["All", "PcG", "trxG"], ['Total context++ score', 'Cumulative weighted context++ score',
       'Aggregate PCT'])
```

```python
def New_Distribution_check(dataframes, boxplot_percentile, groups, colname):
    
    """
    Creates histograms and boxplots from the given dataframe/s

    - dataframes: list of dataframes which distribution you want to check
    - boxplot_percentile: list. Interval € [0,100] in which you want to limit the boxplots (in case of exponential)
    - groups: list. Names of the dataframes you are analycing
    - colname: list. Columns wich the scores are located. Dataframes equivalent scores must share conames
    """
    # Figure definition
    fig_size = plt.rcParams["figure.figsize"] = [20,20]
    n_score = len(colname) #nº score
    n_dataframes = len(dataframes)

    # Recolector/plotting loop
    for n_s in range(n_score): #loop through scores  
        scores = []
        perc_scores = []
        
        for n_df in range(n_dataframes): #loop through dataframes  
            ## score recolector
            df = dataframes[n_df]
            score = df[colname[n_s]].dropna().values
            ### general/hist
            positive = False if min(score)<0 else True  #pythonic if/else 
            scores.append(np.log(abs(score)+1))
            ### boxplot
            raw_s = df[colname[n_s]].dropna().values
            perc = np.percentile(raw_s, boxplot_percentile)
            perc_scores.append(raw_s[(raw_s>perc[0])*(raw_s<perc[1])])

            
        ## Plot
        ### hist
        plt.subplot(n_score, 2 , n_s*2+1)
        if positive == False:
            scores = [ -x for x in scores] #change score sign
            plt.hist(scores, label=groups, bins=50, density=True)
            plt.xticks(plt.xticks()[0],-np.around(np.exp(-plt.xticks()[0])-1,2))
        else: 
            plt.hist(scores, label=groups, bins=50, density=True)
            plt.xticks(plt.xticks()[0],np.around(np.exp(plt.xticks()[0])-1,2))
        plt.legend(loc = "best")
        
        ### boxplot
        plt.subplot(n_score, 2 , n_s*2+2)
        plt.boxplot(perc_scores, labels=groups)



    
        
New_Distribution_check(dataframes=[miR_targ, miRtarg_PcG, miRtarg_trxG],
                       boxplot_percentile=[1,99],
                       groups=["All", "PcG", "trxG"],
                       colname=['Total context++ score', 'Cumulative weighted context++ score','Aggregate PCT'])
```
